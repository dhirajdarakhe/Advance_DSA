```python
# Importing required libraries
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix, classification_report
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing dataset
df = pd.read_csv("https://raw.githubusercontent.com/shivang98/Social-Network-ads-Boost/master/Social_Network_Ads.csv")
df
```

```python
X = df[["Age", "EstimatedSalary"]]
Y = df["Purchased"]

print("X: \n", X)
print("\nY: \n", Y)
```

```python
# Splitting the dataset into training and testing sets (75% training, 25% testing)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
pred = classifier.predict(X_test)

```

```python
cm = confusion_matrix(Y_test, pred)
print(cm)
```

```python
tn = cm[1][1]
tp = cm[0][0]
fn = cm[0][1]
fp = cm[1][0]
print("TP:", tp)
print("FP:", fp)
print("TN:", tn)
print("FN:", fn)
```

```python
precision = tp / (tp+fp)
print(precision)
```

```python
recall = tp / (tp + fn)
print(recall)
```

```python
acc = tp + tn / (tp+tn+fp+fn)
acc = acc/100
print(acc)
```

```python
error_rate = 1 - acc
# print(error_rate)
# print("Error Rate= %0.3f" %error_rate)
print("Error Rate = %0.3f"  %error_rate)
```
