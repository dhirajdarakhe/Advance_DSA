1. Use the inbuilt dataset 'titanic' as used . Plot a box plot for distribution of age with respect to
each gender along with the information about whether they survived or not. (Column names:
'sex' and 'age')
2. Write observations on the inference from the above statistics.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.boxplot(x='age',y="sex",hue='survived',data=df)
plt.title('Distribution of Age with Respect to Gender and Survival')
plt.show()
```
