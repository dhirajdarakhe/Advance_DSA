```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = sns.load_dataset('iris')
df2 = df1.copy()
```

```python
df3 = df1.groupby(by='species')
```

```python
df3.mean()
```

```python
# Percentile distribution of the 3 species

labels = ['setosa', 'versicolor', 'virginica']
size = df2['species'].value_counts()
plt.pie(size, labels = labels, shadow = False, autopct = '%.2f%%')
plt.title('Iris Species', fontsize = 20)
plt.legend()
plt.show()
```

```python
# species				
l  = ["setosa",	"versicolor",	"virginica"	]
size = df2['species'].value_counts()
plt.pie(size, labels=l, shadow=False, autopct='%.02f%%')
plt.title('Iris Species')
plt.show()
```

```python
col = df2.select_dtypes(include=['int64', 'float64']).columns.tolist()
for c in col:
   print(f'Column name : {c}')
   print(f'Column mean: {df2[c].mean()}')
```
