# Text Analytics

1. Extract Sample document and apply following document preprocessing 
methods: Tokenization, POS Tagging, stop words removal, Stemming and 
Lemmatization.
2. Create representation of document by calculating Term Frequency and Inverse 
Document Frequency.


```python
import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
```

```python
import string, nltk
```

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
text1 = '''Data science is an interdisciplinary academic field that uses statistics, scientific computing, scientific methods, processes, algorithms and systems to extract or extrapolate knowledge and insights from noisy,structured, and unstructured data. Data science also integrates domain knowledge from the underlying applicationdomain (e.g., natural sciences, information technology, and medicine). Data science is multifaceted and can bedescribed as a science, a research paradigm, a research method, a discipline, a workflow, and a profession.'''

text2 = '''Data science is the study of data to extract meaningful insights for business. It is a multidisciplinary approach that combines principles and practices from the fields of mathematics, statistics, artificial intelligence, and computer engineering to analyze large amounts of data.'''
```

```python
nltk.download('punkt')
token1 = nltk.word_tokenize(text1)
token2 = nltk.word_tokenize(text2)
```

```python
print(token1)
```

```python
nwtoken = []
for i in token1:
  if i not in [',','.','(',')']:
    nwtoken.append(i)
token1 = nwtoken
```

```python
nwtoken = []
for i in token2:
  if i not in [',','.','(',')']:
    nwtoken.append(i)
token2 = nwtoken
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.pos_tag(token1)
nltk.pos_tag(token2)
```

```python
nltk.download('stopwords')
sw = set(stopwords.words('english'))
```

```python
print(sw)
```

```python
rsw = []
for i in token1:
  if i not in sw:
    rsw.append(i)
token1 = rsw
```

```python
rsw = []
for i in token1:
  if i not in sw:
    rsw.append(i)
token1 = rsw
```

```python
ps = PorterStemmer()
```

```python
for w in token1:
  print(w+" - > ",ps.stem(w))
```

```python
lm = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
for w in token1:
  print(w+" - > ",lm.lemmatize(w))
```

```python
wcnt=0
len=0
for i in token1:
    if i in ['Data']:
        wcnt+=1
    len+=1
print("TF - ",wcnt/len)
```

```python
#No of document(N)=2
import math
N=2
cnt=0
for i in token1:
    if(i=='academic'):
        cnt+=1
        break
for i in token2:
    if(i=='academic'):
        cnt+=1
        break
print("ITF - ",math.log(N/cnt))
```
